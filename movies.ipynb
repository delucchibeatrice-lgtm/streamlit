### Data Cleaning

```python
df['year'] = df['title'].str.extract(r'\((\d{4})\)')
df['title'] = df['title'].str.replace(r'\s*\(\d{4}\)', '', regex=True)
df.head()
```

```python
df.to_csv("movies_cleaned.csv", index=False)
```
